# Case 3: Top 5 negara dengan imigrasi terbanyak ke Canda selama 1980-2013

```python
# buat dataframe baru yang disortir rows menurut data kolom total, dari besar
#ke kecil (descending)
df_can_sorted = df_can.sort_values(by= 'Total', ascending=False, axis=0)

# ambil 5 baris pertama saja untuk jadi top 5 
df_top5 = df_can_sorted.head(5)

# ambil kolom yang years saja yaitu 1980-2013
df_top5_years = df_top5[years]

# kita transpose sehingga indexnya adalah years 1980-2013 
df_top5_years = df_top5_years.T

# di plot, index menjadi x-axis 
df_top5_years.plot()
```

```python
# karena plotnya kecil sekali, kita tambahkan opsi figsize"figure size"
df_top5_years.plot(figsize=(15,10))

# belum selesai, kita perlu berikan judul dan label axisnya
plt.title("5 Negara dengan Imigrasi Terbanyak ke Canada  selama 1980-2013")
plt.ylabel("Jumlah Imigran")
plt.xlabel("Tahun")

plt.show()
```

## Tipe plot

    sejauh ini tipe plotnya adalah default kind, yaitu line 
    tetapi ada banyak tipe plot lainnya, contohnya tipe area

Area plot

```python
# membuat area plot, cukup tambahkan opsi kind = area 
df_top5_years.plot(kind="area", figsize=(12,6))
```

```python
# kalau tidak mau datanya ditumpuk, kita tambahkan opsi stacked=false 
df_top5_years.plot(kind="area", stacked=False, figsize=(12,6))
```

```python
# jangan lupa tambahkan label ke hasil plotnnya
df_top5_years.plot(kind="area", stacked=False, figsize=(12,6))

plt.title("Top 5 Immigration to Canada 1980-2013")
plt.ylabel("Number of immigration")
plt.xlabel("Years")

plt.show()
```

```python
# perhatikan kalau warna arenya agak agak transparan. 
# seberapa transparan warnanya bisa kita atur dengan opsi alpha (0% tidak kel
#hatan, 100% tidak transparan)
df_top5_years.plot(kind="area", stacked=False, figsize=(12,6), alpha=0.3)
```

# Teknik Scripting layer VS Artist layer

#### Scripting layer 
teknik menggunakan plt untuk menambahkan text ke plot yang terakhir kita buat 

#### Artist layer
teknik memasukkan plot kita ke dalam sebuah variabel yang kemudian kita
panggil untuk set_title, atau set jenis text lainnya

```python
ax = df_top5_years.plot(kind="area", stacked=False, figsize=(12,6))
ax.set_title("Top 5 Immigration to Canada 1980-2013")
ax.set_ylabel("Number of Imigrants")
ax.set_xlabel("Years")
```

# Case 7: Histogram-frekuensi distribusi data 

```python
# ambil datanya dulu untuk denmark, norway, sweden pada kolom taun 1980-2013
df_dns = df_can.loc[["Denmark","Sweden","Norway"], years]
df_dns
```

```python
# kalau langsung kita plot 
df_dns.plot(kind="hist")
```

```python
# bentuknya masih salah 
df_dns = df_dns.T
df_dns.plot(kind="hist")
```

```python
# kasih label 
df_dns.plot(kind="hist")
plt.title("Histogram imigrasi dari denmark, sweden, norwat ke canada pada tahun 1980-2013")
plt.ylabel("jumlah tahun")
plt.xlabel("jumlah imigran")
```

```python
# garis xlabelnya tidak sesuai jumlahnya 
# kita minta bantuan numpy 
count, bin_edges = np.histogram(df_dns)

print(count)
print(bin_edges)
```

```python
# secara default np.histogram akan memberikan 10 bins 
# kita bisa set untuk minta jumlah bins berbeda

count, bin_edges= np.histogram(df_dns, 15)

print(count)
print(bin_edges)
```

```python
# berikan jumlah bins baru ke plot kita 
df_dns.plot(kind="hist", xticks=bin_edges, bins=15)
plt.title("Histogram imigrasi dari denmark, sweden, norwat ke canada pada tahun 1980-2013")
plt.ylabel("jumlah tahun")
plt.xlabel("jumlah imigran")
```

```python
# berikutnya kita edit ukurna plotnya 
# juga warnya bin nya 
# dan transparasi warnanya 
df_dns.plot(kind="hist", bins=15, xticks=bin_edges, figsize=(20,10), color=
           ['Coral', 'darkslateblue', 'mediumseagreen'], alpha=0.6)

plt.title("Histogram imigrasi dari denmark, sweden, norwat ke canada pada tahun 1980-2013")
plt.ylabel("jumlah tahun")
plt.xlabel("jumlah imigran")
plt.show()
```

# Pie Chart 

sebelum membuat pie chart, kita siapkan dulu data dengan presentasenya. karena 
pie chart bukan untuk plot data excel dengan x dan y axis, tapi lebih ke kumpulan kategori, masing masing kategori itu berapa persen dari keseluruhan 

di contoh ini, kita akan mencari presentase imigran per benua 

```python
# benua2 ini presentase mengirim imigran ke Canadanya masing2 berapa persen
df_continent = df_can.groupby('Continent', axis=0).sum()
df_continent
```

```python
# kita ambil dari data total saja baru kita plot 
df_continent['Total'].plot(kind="pie")
```

```python
df_continent['Total'].plot(
    kind="pie",
    figsize=(5,6),
    autopct="%1.1f%%",
    startangle=0
)
```

```python
df_continent['Total'].plot(
    kind="pie",
    figsize=(15,6),
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen',
           'pink'],
    explode=[0.1,0,0,0,00.1,0.1],
    pctdistance=1.12,
    labels=None
)

plt.title("Total Immigrants to Canada 1980-2013 by Continent", y=1.12)

# memastikan size graph nya sesuai figsize 
plt.axis('equal')

# tambahin legent menggunakan script
plt.legend(labels=df_continent.index, loc="upper left")

plt.show()
```
